### Problem 1:

1. Write a program to calculate area of shapes. Your program should be capable of calculating the area of a square, rectangle, triangle and a circle. The user should be presented with options to select the shape. Based on which shape is chosed by the user, the program should ask for the appropriate input and print the resulting area on the screen.

When the program is run, the screen should display something like this:

**Which shape would you like to calculate the area for? Please enter the option number-**<br>
**1. Square**<br>
**2. Rectangle**<br>
**3. Triangle**<br>
**4. Circle**<br>
**Enter Option: _**

Say the user enters the option 1.

**Please enter the length of a side: _**

If the user enters a value of 5. The output should be:

**The area of the square is 25**

This program should indicate that the input is invalid if the user enters a character instead of a number as input. For instance if the user enters a value of ‘a’ instead of 5 in the previous example the program should prompt:

**Invalid input, please enter a number: _**


```python
# PLEASE RUN THE NEXT CELL BEFORE PROCEEDING WITH THE PROGRAM

print('Which shape would you like to calculate the area for ?\n 1. Square\n 2. Rectangle\n 3. Triangle\n 4. Circle\n')
print('Enter 1 for Square, 2 for Rectangle, 3 for Triangle or 4 for Circle !!')
option = input('Enter Option: ')
final_option = int(validate_ip_for_shape(option))


if final_option == 1:
    area_of_square()
elif final_option == 2:
    area_of_rectangle()
elif final_option == 3:
    area_of_triangle()
else:
    area_of_circle()
```

```python
def validate_ip_for_shape(choice):
    if choice.isalpha():
        choice = input('Invalid input, please enter a number:')
        return validate_ip_for_shape(choice)
    elif choice in ('1','2','3','4'):
        return choice
    else:
        choice = input('Invalid input, please enter a number:')
        return validate_ip_for_shape(choice)
    
def validate_ip_for_side(choice):
    if choice.isalpha():
        choice = input('Invalid input, please enter a number:')
        return validate_ip_for_side(choice)
    elif choice.isdigit() == False:
        choice = input('Invalid input, please enter a number:')
        return validate_ip_for_side(choice)  
    elif choice == '0':
        choice = input('Invalid input, please enter a number:')
        return validate_ip_for_side(choice)
    else:
        return choice
    
def area_of_square():
    side = input('Enter the side of the square in cms: ')
    final_side = float(validate_ip_for_side(side))
    print('Area of square = {0} sq.cms '.format(final_side*final_side))
    
def area_of_rectangle():
    length = input('Enter the length of rectangle in cms:')
    final_length = float(validate_ip_for_side(length))
    breadth = input('Enter the breadth of rectangle in cms: ')
    final_breadth = float(validate_ip_for_side(breadth))
    print('Area of rectangle = {0} sq.cms '.format(final_length*final_breadth))
    
def area_of_triangle():
    height = input('Enter the height of triangle in cms: ')
    final_height = float(validate_ip_for_side(height))
    base = input('Enter the base of triangle in cms:')
    final_base = float(validate_ip_for_side(base))
    print('Area of triangle = {0} sq.cms '.format(0.5*final_height*final_base))

def area_of_circle():
    rad = input('Enter the radius of circle:')
    final_rad = float(validate_ip_for_side(rad))
    print('Area of circle = {0} sq.cms '.format(3.14*final_rad*final_rad))
```

### Problem 2:

Write a function that would require 2 arguments – the input text to be encrypted and a key. For eg: Given the input text ‘hello’ and the key 3, the resulting encryted text would be ‘khoor’. Here you can see that every character in the string hello is shifted by 3 characters. ‘h’ has shifted to ‘k’, ‘e’ has shifted to ‘h’ and so on. If a key of 5 were used, the resulting string would be ‘mjqqt’. This function should be capable of ignoring any characters which are not alphabets. Th2 character ‘z’ entered b
y the user for a key of 3 would result in ‘c’.

Usage:<br>

Similarly create decrypter which can decode the encryted text when provided the input text and key

Usage:<br>
**decrypt(‘khoor zruog!’, 3)**<br>

For the sake of simplicity you can assume that input solely consists of lowercase alphabets, spaces and punctuation symbols. Numbers in the input text would also be ignored similar to symbols.

```python
# PLEASE RUN THE NEXT CELL BEFORE PROCEEDING WITH THE PROGRAM

print('1 : Encryption \n2: Decryption')
print('Enter 1 for Encryption or 2 for Decryption')

option = input('Enter your choice :')
validated_option = int(validate_choice(option))

if validated_option == 1:
    ptext = input('Enter the text to be encrypted: ')
    encrpty_key = input('Enter the key:')
    final_encrypt_key = int(validate_key(encrpty_key))
    encrypt(ptext,final_encrypt_key)

elif validated_option == 2:
    ctext = input('Enter the text to be decrypted: ')
    decrpty_key = input('Enter the key:')
    final_decrypt_key = int(validate_key(decrpty_key))
    decrypt(ctext,final_decrypt_key)
```

```python
def encrypt(string,key):
    string = string.strip()
    string = string.lower()
    cipher_text = ''
    for i in string:
        if i in ' !@#$%^&*()""0123456789':
            cipher_text = cipher_text + i
        else:
            pos_i = ord(i)
            old_index = ord(i) - ord('a')
            new_index = (old_index + key) % 26
            new_i = new_index + ord('a')
            new_char = chr(new_i)
            cipher_text = cipher_text + new_char
    print('Original text =',string)
    print('Encrypted text =',cipher_text)

    
def decrypt(cipher_text,key):
    cipher_text = cipher_text.strip()
    cipher_text = cipher_text.lower()
    string = ''
    for i in cipher_text:
        if i in ' !@#$%^&*()""0123456789':
            string = string + i
        else:
            pos_i = ord(i)
            old_index = ord(i) - ord('a')
            new_index = (old_index - key) % 26
            new_i = new_index + ord('a')
            new_char = chr(new_i)
            string = string + new_char
    print('Encrypted text =',cipher_text)
    print('Original text =',string)
    
    
def validate_key(key):
    if key.isalpha() or key == '' or key== ' ':
        key = input('Invalid key, please enter again: ')
        return validate_key(key)
    else:
        return key
    
    
def validate_choice(choice):
    if choice.isalpha():
        choice = input('Invalid input, please enter a number:')
        return validate_choice(choice)
    elif choice in ('1','2'):
        return choice
    else:
        choice = input('Invalid input, please enter a number:')
        return validate_choice(choice)
```
